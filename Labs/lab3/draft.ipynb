```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# global variables
filename = "unlabelled_data.csv"
```

### a) Read file and scatter plot
### b) Draw the line

```python
def read_file(filename):
    
    df = pd.read_csv(filename, header = None)
    df.columns = ["x", "y"]
    
    return df


dataframe = read_file(filename)
print(dataframe.iloc[1][1])
print(dataframe)
```

```python
def scatter_points(df):

    # plot dataframe
    plt.scatter(df["x"], df["y"], color = "aquamarine", edgecolors = "mediumaquamarine", label = "data points", linewidth = 1, s = 25, alpha = 0.75, zorder = 3)
    plt.axis("equal")
    plt.axhline(y = 0, color = "lightseagreen", linewidth = 0.5, zorder = 2)
    plt.axvline(x = 0, color = "lightseagreen", linewidth = 0.5, zorder = 2)
    max_limit = max(abs(dataframe.values.min()), abs(dataframe.values.max()))   # dessa rader fick jag frn Claude, behver dubbelkollas. gr att plot center r origo
    plt.xlim(-max_limit, max_limit)                                             # dessa rader fick jag frn Claude, behver dubbelkollas. gr att plot center r origo
    plt.ylim(-max_limit, max_limit)                                             # dessa rader fick jag frn Claude, behver dubbelkollas. gr att plot center r origo
    plt.title("A scatter plot of the dataframe")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.grid(True, color="gainsboro", linestyle="dashed", zorder=1)

    # draw line
    k = -1
    m = 0
    x = np.linspace(-10, 10, 600)
    y = k * x + m
    plt.plot(x, y, color = "lightseagreen", label = ("regression line"))
    plt.legend(loc = "upper left")

    return x, y, k, m


X, Y, K, M = scatter_points(dataframe) 
```
