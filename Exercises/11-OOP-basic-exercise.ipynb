</details>


````python
class Convert(object):

    """ 
    1 foot (US) ≈ 0.3048 meters (EU)
    1 inch (US) ≈ 2.5400 centimeters (EU)
    1 pound(US) ≈ 0.4536 kilograms (EU)
    """
    
    def __init__(self, value):
        if not isinstance(value, (float, int)):
                raise TypeError(f"{value} is not a valid number.")
        if value <= 0:
             raise ValueError(f"{value} is equal to or lower than zero.")
        self._value = value
    
    @property
    def foot_to_m(self):
        return (self._value * 0.3048)
    
    @property
    def inch_to_cm(self):
        return (self._value * 2.5400)
    
    @property
    def pound_to_kg(self):
        return (self._value * 0.4536)
    
    @property
    def __repr__(self):
        pass


feet = 7
meters = Convert(feet)
print(f"{feet} feet is {meters.foot_to_m} meters")

inches = 12
centimeters = Convert(inches)
print(f"{inches} inches is {centimeters.inch_to_cm} centimeters")

pounds = 150
kilograms = Convert(pounds)
print(f"{pounds} pounds is {kilograms.pound_to_kg} kilograms")

shoe_sizes_US = [7, 6.5, 5, 8.5, 10]
# feet_sizes_EU = [unit.inch_to_cm (unit=Convert(value)) for value in shoe_sizes_US]
feet_sizes_EU = [Convert(value).inch_to_cm for value in shoe_sizes_US]
# feet_sizes_EU = []
# for value in shoe_sizes_US:
#      unit = Convert(value)
#      feet_sizes_EU.append(unit.inch_to_cm) 
print()
print(f"A list comprehension can both create and convert.\nOriginal list (inches): {shoe_sizes_US}\nConverted list (cm): {feet_sizes_EU}")
````
